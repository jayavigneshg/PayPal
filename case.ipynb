```python
import pandas as pd
from matplotlib import pyplot as plt
df_pp=pd.DataFrame(pd.read_csv("C:\\Users\\jayavign\\Documents\\Personal\\PayPal\\data.csv"))
print(df_pp.describe())
plt.hist(df_pp["score"])
plt.show() 
```

```python
# df_pp
df_ari=df_pp[df_pp["name"] =='Arielle']
print(df_ari.describe())
plt.hist(df_ari["score"])
plt.show() 
```

```python
is_righthand=df_ari.hand=='R'
df_ari_rh=df_ari[is_righthand]
print(df_ari_rh.describe())
plt.hist(df_ari_rh["score"])
plt.show() 
```

```python
is_lefthand=df_ari.hand=='L'
df_ari_lh=df_ari[is_lefthand]
print(df_ari_lh.describe())
plt.hist(df_ari_lh["score"])
plt.show() 
```

```python
df_bor=df_pp[df_pp["name"] =='Boris']
print(df_bor.describe())
plt.hist(df_bor["score"])
plt.show() 
```
